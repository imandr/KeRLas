```python
import matplotlib
%matplotlib inline
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
data = {}
with open("tmp/monitor.csv", "r") as f:
    csv_reader = csv.reader(f)
    labels = None
    first = True
    for row in csv_reader:
        if first:
            labels = row
            first = False
        else:
            for label, word in zip(labels, row):
                col = data.setdefault(label, [])
                col.append(float(word) if not word in ('', None) else None)
tcol = data['t']
del data['t']
#test_scores = data['test_score']
#del data['test_score']

fig, axs = plt.subplots(2, 1, figsize=(15,10))
ax1, ax2 = axs
for label, col in data.items():
    if label != 't' and label != "score":
        ax1.plot(tcol, col)
ax1.grid()
#ax2.plot(tcol, test_scores)
#ax2.grid()
```
